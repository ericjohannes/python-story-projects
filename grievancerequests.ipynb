```python
#all categories with dates, I misspelled date as data,oh well
data_cats = ["OG_INCIDENT_DT", "OG_TORTE_CLAIM_DT", "OG_APPEAL_REC_ACKN_DT", "OG_SUB_COA_DT", "OG_ROLE_ASSIGNED_DATE",
            "OG_STAFF_DETERMIN_DT", "OG_FORM_ISSUE_DT", "OG_LEVEL_REC_DT",]
```

```python
#change into datetime formats
for cat in data_cats:
    mydata[cat] = pd.to_datetime(mydata[cat], format='%d%b%Y:%H:%M:%S', 
                                          errors='coerce')
```

```python
grouped_year = mydata.groupby(mydata.OG_INCIDENT_DT.dt.year).count()
```

```python
# shows total number of incidents by year
grouped_year.SF_FACILITY_ID.plot()
#plt.show()
plt.title("All Facilities")
plt.xlabel('year')
plt.ylabel('Number of grievances')
plt.savefig(fname="figures/"+"all facilitiess")
```

```python
topics = mydata.GEN_TOPIC.unique()
print(topics)
```

```python
cleantopics = []
```

```python
for topic in topics:
    newtopic = topic.replace('/'," or ")
    newtopic = newtopic.replace('.','')
    newtopic = newtopic.replace(',','')
    newtopic = newtopic.replace('-','')
    cleantopics.append(newtopic)

print(cleantopics)
```

```python
plt.clf()
```

```python
range(0,len(topics))
```

```python
# for each topic
# slice out of mydata rows where topic matches
# group by year
# make plot of count of fac id

for i in range(0,len(topics)):
    onetopic_df = mydata[mydata.GEN_TOPIC == topics[i]]
    onetopic_byyear = onetopic_df.groupby(mydata.OG_INCIDENT_DT.dt.year).count()
    onetopic_byyear.SF_FACILITY_ID.plot()
    plt.title("Topic: " + topics[i])
    plt.xlabel('year')
    plt.ylabel('Number of grievances')
    #plt.show()
    plt.savefig(fname="figures/topics/"+cleantopics[i])
    plt.clf()
```
